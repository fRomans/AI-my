```python
from utils import ChatOpenAI
from getpass import getpass

#course_api_key= "Введите ваш API ключ с курса"
course_api_key = getpass(prompt='Введите API ключ')

# Инициализируем языковую модель
llm = ChatOpenAI(course_api_key=course_api_key)
```

```python
# подгрузим датасет
import pandas as pd

df = pd.read_csv("https://stepik.org/media/attachments/lesson/1084404/dial_df.csv")
df
```

```python
# Импортируйте модуль памяти
from langchain.chains import ConversationChain, LLMChain
from langchain.memory import ConversationBufferMemory


# создайте цепочку диалога, в параметры передайте языковую модель и память
memory = ConversationBufferMemory()
conversation_buf_mem = ConversationChain(
    llm= llm,  # языковая модель
    memory = memory, # модуль памяти
    verbose=True # отображать логи в консоли или нет
)

```

```python

ans = []
for id, q in zip(df['dialogue_id'], df['question']):
    
    # prompt = prompt_template.format(text_input=memory, q=q) # Добавляем сообщение в промпт
    # amount = llm.invoke(prompt).content # Ответ модели
    ans.append(conversation_buf_mem.invoke(q+". Отвечай только цифрами")['response'])
    # break # уберем break, когда убедимся, что работает на одном примере
    
df['answer'] = ans # запишем предсказания в датафрейм
df.to_csv('solution.csv', index=False) # сохраним решение в csv  
conversation_buf_mem.memory.clear()
```
